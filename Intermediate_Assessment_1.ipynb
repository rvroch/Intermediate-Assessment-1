**Distribution of Arrival Delays**

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the NYC Flights dataset
df = pd.read_csv('C:/Users/rocha/Downloads/ny-flights.csv')

# Drop rows with missing arrival delay values
df = df.dropna(subset=['arr_delay'])

# Plotting the distribution of arrival delays
plt.figure(figsize=(12, 6))

# Histogram
plt.subplot(1, 2, 1)
sns.histplot(df['arr_delay'], bins=50, kde=True, color='blue')
plt.title('Distribution of Arrival Delays', fontsize=14)
plt.xlabel('Arrival Delay (minutes)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xlim(-50, 300)  # Adjust x-axis to focus on the most relevant range

# Box Plot
plt.subplot(1, 2, 2)
sns.boxplot(x=df['arr_delay'], color='orange')
plt.title('Box Plot of Arrival Delays', fontsize=14)
plt.xlabel('Arrival Delay (minutes)', fontsize=12)
plt.xlim(-50, 300)  # Adjust x-axis to match the histogram

# Display the plots
plt.tight_layout()
plt.show()
````

## **2. Diabetes Dataset**

### Univariate Analysis

````python
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
df=pd.read_csv("C:/Users/rocha/Downloads/diabetes_model.csv")
df
````

**Descriptive Statistics**

````python
df.describe()
````

**Visualizations**

````python
sns.histplot(df['Glucose'], kde=True)
plt.title('Distribution of Glucose')
plt.show()
````

````python
sns.boxplot(x=df['BMI'])
plt.title('Boxplot of BMI')
plt.show()
````

````python
outcome_counts = df['Outcome'].value_counts()
plt.pie(outcome_counts, labels=['Non-Diabetic', 'Diabetic'], autopct='%1.1f%%', startangle=90)
plt.title('Diabetic vs Non-Diabetic Percentage')
plt.show()
````

### Bivariate Analysis

**Scatter Plot**

````python
sns.scatterplot(x='Glucose', y='BMI', data=df, hue='Outcome')
plt.title('Glucose vs BMI')
plt.xlabel('Glucose')
plt.ylabel('BMI')
plt.show()
````

**Box Plot**

````python
sns.boxplot(x='Outcome', y='Age', data=df)
plt.title('Age Distribution by Outcome')
plt.show()
````

````python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
````

### Multivariate Analysis

**Pair Plot**

````python
sns.pairplot(df[['Pregnancies', 'Glucose', 'BMI', 'Age', 'Outcome']], hue='Outcome')
plt.show()
````

**Correlation Matrix**

````python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
````

**Stacked Bar Plot**

````python
outcome_by_pregnancies = df.groupby('Pregnancies')['Outcome'].value_counts().unstack().fillna(0)
outcome_by_pregnancies.plot(kind='bar', stacked=True, colormap='Set1')
plt.title('Outcome by Number of Pregnancies')
plt.xlabel('Number of Pregnancies')
plt.ylabel('Count')
plt.show()
````

**Heatmap**

````python
outcome_by_pregnancy_age = pd.crosstab(df['Pregnancies'], df['Outcome'])
sns.heatmap(outcome_by_pregnancy_age, annot=True, cmap='Blues', fmt='d')
plt.title('Outcome by Number of Pregnancies')
plt.show()
````
